製作圖表，顯示113年每個月的月均氣溫和濕度

```python
import plotly.graph_objects as go
import pandas as pd

months = df_113y['month']
temperature = df_113y['temperature']
humidity = df_113y['humidity']

# 確保 temperature 和 humidity 是數值型
df_113y['temperature'] = pd.to_numeric(df_113y['temperature'], errors='coerce')
df_113y['humidity'] = pd.to_numeric(df_113y['humidity'], errors='coerce')

fig = go.Figure()
fig.add_trace(go.Bar(
    x=months,
    y=temperature,
    name='Temperature',
    marker_color='indianred'
))
fig.add_trace(go.Bar(
    x=months,
    y=humidity,
    name='Humidity',
    marker_color='lightsalmon'
))

fig.update_layout(
    barmode='group',
    title='Taiwan 113y Monthly Temperature and Humidity',
    xaxis_title="Monthly", yaxis_title="Temperature And Humidity",
    legend_title="Legend",
    title_x=0.5,
    yaxis=dict(
        range=[0, 100],  # 顯示範圍 0～100
        dtick=20,        # 每 20 為一格
        tickmode='linear',  # 線性刻度（避免自動調整）
    )
)
fig.show()
```

台灣101年至114年，每年每月的降雨量



```python
import plotly.express as px

# 轉換數據類型
df.iloc[:, 1:] = df.iloc[:, 1:].apply(pd.to_numeric)

# 確保 'month' 欄位是按數字順序的分類型資料
df['month'] = pd.Categorical(df['month'], categories=list(range(1,13)), ordered=True)

# 依照年和月排序資料
df = df.sort_values(by=['year', 'month'])

month_colors = {
    1: '#4B6CB7',  # 冬 - 冰藍
    2: '#6A4C93',  # 冬 - 暗紫
    3: '#77DD77',  # 春 - 嫩綠
    4: '#FFB7C5',  # 春 - 粉紅
    5: '#A1C4FD',  # 春 - 淺藍
    6: '#FFD166',  # 夏 - 豔黃
    7: '#FF6B6B',  # 夏 - 西瓜紅
    8: '#FFA07A',  # 夏 - 淺橘
    9: '#D4A373',  # 秋 - 焦糖棕
    10: '#BDB76B', # 秋 - 卡其黃
    11: '#8D99AE', # 秋 - 淡灰藍
    12: '#1E90FF'  # 冬 - 亮藍
}

df['year'] = df['year'].astype(str)
# Plot using Plotly
fig = px.bar(df, x='year', y='precipitation', color='month',
             title="Taiwan Monthly Precipitation Per Year",
             labels={'precipitation':'Precipitation(mm)','month':'Month'},
             barmode='stack',
             color_discrete_map=month_colors)

# Update layout for better visualization
fig.update_layout(xaxis_title="Year", yaxis_title="Precipitation(mm)",
                  legend_title="Month", title_x=0.5)

# Show the plot
fig.show()
```

```python
# 計算每年降雨量總和
yearly_precipitation_sum = df.groupby('year')['precipitation'].sum().reset_index()
monthly_precipitation_sum = df.groupby('month')['precipitation'].sum().reset_index()

# 印出結果
print(yearly_precipitation_sum)
```

```python
import plotly.express as px
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# 篩選出 113 年資料
df_113y = df[df['year'] == 113].copy()

# 確保欄位是數值型別
df_113y['precipitation'] = pd.to_numeric(df_113y['precipitation'], errors='coerce')
df_113y['precipitation_days'] = pd.to_numeric(df_113y['precipitation_days'], errors='coerce')

# 使用KMeans進行聚類
kmeans = KMeans(n_clusters=4)
df_113y['cluster'] = kmeans.fit_predict(df_113y[['precipitation', 'precipitation_days']])

# 使用PCA進行降維視覺化
pca = PCA(n_components=2)
components = pca.fit_transform(df_113y[['precipitation', 'precipitation_days']])
df_113y['pca1'] = components[:, 0]
df_113y['pca2'] = components[:, 1]

# 視覺化聚類結果
fig = px.scatter(
    df_113y,
    x='pca1',
    y='pca2',
    color='cluster',
    title="年度降雨聚類結果",
    labels={'pca1': '主成分1', 'pca2': '主成分2', 'year': '群組'},
    hover_data=['month', 'precipitation', 'precipitation_days']  # 在互動中顯示更多細節
)

# 更新圖表的佈局
fig.update_layout(
    xaxis_title="主成分1",
    yaxis_title="主成分2",
    legend_title="群組",
    title_x=0.5,  # 圖表標題居中
    height=600,
    width=800
)

# 顯示圖表
fig.show()
```
