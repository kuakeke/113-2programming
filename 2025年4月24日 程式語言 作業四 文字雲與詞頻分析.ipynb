```python
import jieba
from snownlp import SnowNLP
import pandas as pd
from collections import Counter
import re
import plotly.express as px
from wordcloud import WordCloud
```

```python
# 把所有內容欄位的文字合併成一整段
text = ' '.join(dicts['內容'].tolist())
print(text)
```

```python
# 去除標點符號
word = re.sub(r'[^\w\s]', '', text)

# 使用 jieba 進行斷詞
words = jieba.lcut(word)

# 去除無意義詞（停用詞）
stopwords = set([
    '雖然', '的', '了', '和', '是', '我', '有', '就', '不', '都', '上', '也', '很',
    '到', '說', '要', '去', '你', '會', '著', '沒有', '看', '好', '自己', '這'
])

filtered_words = [word for word in words if word not in stopwords and len(word.strip()) > 1]

# 統計詞頻
from collections import Counter
word_freq = Counter(filtered_words)

# 印出前 10 名高頻詞
print(word_freq.most_common(10))
```

```python
from collections import Counter
import pandas as pd

word_freq = Counter(filtered_words)

# 轉成 DataFrame
word_freq_df = pd.DataFrame(word_freq.items(), columns=['Word', 'Frequency'])

# 排序
word_freq_df = word_freq_df.sort_values(by='Frequency', ascending=False).reset_index(drop=True)
```

```python
import plotly.express as px

fig = px.bar(word_freq_df.head(20), x='Word', y='Frequency',
             title='關鍵字出現次數統計（前20名）',
             labels={'Word': '關鍵字', 'Frequency': '出現次數'},
             text='Frequency')

fig.update_layout(
    xaxis_title="關鍵字",
    yaxis_title="出現次數",
    xaxis=dict(showgrid=True),
    yaxis=dict(showgrid=True),
    template='plotly_white'
)

fig.show()
```

**從詞頻圖可以看見，這5篇作文《我的家人》，使用最多的詞是“爸爸、媽媽、哥哥”。如果是一位老師透過這個程式，可快速看到每位同學寫關於《我的家人》的作文，提及最多的家庭成員是誰。**

```python
# 使用 WQY Zenhei 字體
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    font_path='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc'
)

# 根據詞頻生成文字雲
wordcloud.generate_from_frequencies(word_freq)

# 將文字雲轉換為圖像
wordcloud_img = wordcloud.to_array()

# 使用 Plotly 顯示文字雲
fig = px.imshow(wordcloud_img)
fig.update_layout(
    title="文字雲",
    xaxis=dict(showgrid=False, showticklabels=False, zeroline=False),
    yaxis=dict(showgrid=False, showticklabels=False, zeroline=False),
    coloraxis_showscale=False
)
fig.show()
```

**透過雲文字圖，也可明顯看出小學生喜歡使用的詞，也能檢視大致作文內容。例如出現“出國”、“加拿大”，可以看出學生有出國的經驗。或是“唱歌”、“吃糖”，可看出學生喜好。**

```python
import matplotlib.pyplot as plt
import string

# 選取前兩篇作文
selected_df = df.head(5)

# 停用詞
stopwords = set(['的', '了', '和', '是', '我', '也', '有', '就', '不', '在', '都', '很', '到', '說', '一個', '我們', '你', '啊'])

# 清理與斷詞函數
def clean_and_tokenize(text):
    # 確保是字串型別
    text = str(text)
    # 去除標點與空格
    text = ''.join([char for char in text if char not in string.punctuation + '，。、！？；：「」『』（）～…—\n\r '])
    # 使用 jieba 斷詞
    words = jieba.lcut(text)
    # 過濾短詞、停用詞、數字
    return [word for word in words if word not in stopwords and len(word) > 1 and not word.isdigit()]

# 每篇一張文字雲
for idx, row in selected_df.iterrows():
    text = str(row['內容'])  # 請確保這是你的欄位名
    words = clean_and_tokenize(text)
    freq = Counter(words)

    # 濾除非字串 key（避免 str/int 混用錯誤）
    freq = {k: v for k, v in freq.items() if isinstance(k, str)}

    # 製作文字雲
    wc = WordCloud(
     width=800,
     height=400,
      background_color='white',
      font_path='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',
     ).generate_from_frequencies(freq)

    # 將圖片轉為陣列
    wordcloud_img = wc.to_array()

    # 用 plotly.express 顯示
    fig = px.imshow(wordcloud_img)
    fig.update_layout(
        title=f"第 {idx + 1} 篇文字雲",
        xaxis=dict(showgrid=False, showticklabels=False, zeroline=False),
        yaxis=dict(showgrid=False, showticklabels=False, zeroline=False),
        coloraxis_showscale=False
    )
    fig.show()
```

**這是我再針對5篇作文，各別生成的文字雲圖。可以各別檢視單篇作文所使用的詞。第一篇出現“吃糖”、第二篇出現“動物園”，每篇作文的用詞差異一目了然。**
