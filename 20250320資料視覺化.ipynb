```python
import pandas as pd
import matplotlib.pyplot as plt

# Path to your CSV file
file_path = '20250320swimmingpool.csv'

# Read the CSV file
df = pd.read_csv(file_path, skiprows=1, header=0)

# Display the first few rows of the DataFrame
df
```

```python
Day = df[df['Identity'] == 'Outsiders']
```

```python
Day
```

```python
# 透過 pivot_table 計算每天不同身份的總消費金額
Pivot = df.pivot_table(values="Price", index="Day", columns="Identity", aggfunc="sum", fill_value=0)
```

```python
Pivot
```

```python
# 繪製堆疊長條圖
plt.figure(figsize=(6, 6))  # 設定圖表大小
Pivot.plot(kind='bar', stacked=True, figsize=(6, 6))  # 堆疊長條圖

# 設定標籤與標題
plt.xlabel("Day")  # X 軸標籤
plt.ylabel("Price")  # Y 軸標籤
plt.title("Daily Spending by Identity")  # 圖表標題
plt.legend(title="Identity")  # 圖例
plt.grid(axis="y", linestyle="--", alpha=0.7)  # 只加 Y 軸的網格

# 顯示圖表
plt.show()
```
